##Getting The Crypto Feer and Greed index to improve accuracy as these numbers represents the overall market sentiments


##Data Source : https://alternative.me/crypto/fear-and-greed-index/

```python
import requests
import csv
import pandas as pd

# Replace 'your_url_here' with the actual URL from which you want to fetch the JSON data
url = 'https://api.alternative.me/fng/?limit=1827&date_format=us'
file_name = 'fng.csv'

        # Send an HTTP GET request to the URL
response = requests.get(url)
response.raise_for_status()  # Check for any request errors

        # Get the JSON data from the response
json_data = response.json()
print(json_data["data"])

#         # Convert JSON data to DataFrame using pandas
df = pd.DataFrame(json_data["data"])
time=df['timestamp']
# df.drop(df[["value_classification","time_until_update"]],axis=1)
df_reversed = df.iloc[::-1]
df_reversed.to_csv(file_name, index=False)


fng=df_reversed[['value','timestamp']]
fng.head()



```
